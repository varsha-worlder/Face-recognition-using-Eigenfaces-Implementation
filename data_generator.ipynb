```
import random
from faker import Faker
import pandas as pd
from datetime import datetime, timedelta

# Initialize Faker
fake = Faker(['en_US'])


## Staff Profile generator

def generate_gender():
    return random.choice(['Male','Female'])


def generate_age_group_and_dob():
    # Define age groups and their age ranges
    age_groups = {
        "under 18": (0, 17),
        "18-24": (18, 24),
        "25-34": (25, 34),
        "35-44": (35, 44),
        "45-54": (45, 54),
        "55-60": (55, 60),
        "Over 60": (61, 100)
    }

    # Randomly choose an age group
    age_group = random.choice(list(age_groups.keys()))
    min_age, max_age = age_groups[age_group]

    # Generate a random date of birth based on the selected age group
    today = datetime.today()
    birth_date = today - timedelta(days=random.randint(min_age * 365, max_age * 365))
    return age_group, birth_date.strftime("%Y-%m-%d")

def generate_country_region_id():
    countries = ["JP", "PH", "VN", "CN", "BR"]
    return random.choice(countries)

def generate_japanese_address():
    city = random.choice(["Tokyo", "Osaka", "Kyoto", "Sapporo", "Fukuoka", "Nagoya"])
    area = random.choice(["Shibuya", "Shinjuku", "Minato", "Chiyoda", "Kita", "Chūō"])
    address = f"{random.randint(1, 100)}-{random.randint(1, 20)} {area} {city}, Japan"
    return address


def generate_japanese_language_ability():
    levels = ["Beginner", "Intermediate", "Advanced", "Native"]
    return {
        "Conversation": random.choice(levels),
        "Reading": random.choice(levels),
        "Writing": random.choice(levels),
    }

def generate_employment_type():
    return random.choice([
        "FixedTerm", "IndefiniteTerm"
    ])

def random_email():
    domain = random.choice(["@gmail.com", "@email.com"])
    username = ''.join(random.choices('abcdefghijklmnopqrstuvwxyz0123456789', k=8))
    return username + domain

def generate_preferred_work_schedule():
    return random.choice(["Day Shift", "Night Shift", "Rotating Shift"])

def generate_desired_working_hours():
    # List of possible time slots for desired working hours
    time_slots = [
        "9:00 - 17:00",
        "8:00 - 16:00",
        "10:00 - 18:00",
        "9:00 - 18:00",
        "7:00 - 15:00",
        "11:00 - 19:00"
    ]
    return random.choice(time_slots)

def generate_actual_working_hours():
    # Define a random number of working hours, usually between 7 and 9 hours
    return random.choice([5,6,7, 8, 9])

def generate_holidays():
    # Randomly choose whether they get holidays on weekends, holidays, weekdays, or by shift system
    holiday_types = [
        "Saturdays and Sundays",
        "Holidays (National Holidays)",
        "Weekdays (Monday to Friday)",
        "Shift System (Rotating Days Off)"
    ]
    return random.choice(holiday_types)

def generate_preferred_days_off():
    # List of days to pick from for preferred days off
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

    # Randomly choose one or two days off
    preferred_days = random.sample(days_of_week, k=random.randint(1, 2))
    return ", ".join(preferred_days)

def generate_willing_to_work_overtime():
    return random.choice(["Yes", "No"])

def generate_willing_to_work_on_weekends_or_holidays():
    return random.choice(["Yes", "No"])

def generate_desired_salary_amount():
    # Randomly choose between hourly or monthly salary
    salary_type = random.choice(["hourly", "monthly"])

    if salary_type == "hourly":
        # Random hourly salary between 1000 and 5000 yen
        salary_amount = random.randint(1000, 5000)
        return f"{salary_amount} yen/hour"
    else:
        # Random monthly salary between 200,000 and 1,000,000 yen
        salary_amount = random.randint(200000, 1000000)
        return f"{salary_amount} yen/month"

def generate_desired_start_date():
    # Randomly pick a start date between 1 month and 3 months from today
    today = datetime.today()
    start_date = today + timedelta(days=random.randint(30, 120))  # 30 to 90 days from today
    return start_date.strftime("%Y-%m-%d")

def generate_preferred_working_days():
    # List of all days of the week
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

    # Select at least 4 days randomly
    preferred_days = random.sample(days_of_week, random.randint(2, 7))
    return ", ".join(preferred_days)

def generate_qualifications():
    # List of possible qualifications
    qualifications = [
        "Lift License", "Driving License (AT only OK)", "Driving License (not limited to AT)",
        "Large License", "Sling", "Crane", "Bookkeeping", "Nursing Care Staff Beginner Training (Helper 2nd Grade)",
        "Nursery Teacher", "A Licensed Chef", "Nutritionist",
        "Forklift License", "First Aid Certification", "Project Management Certification", "ITIL Certification",
        "Software Development Certification", "HR Certification", "Business Analysis Certification", "Certified Nurse Assistant"
    ]

    # Select a few qualifications randomly
    selected_qualifications = random.sample(qualifications, random.randint(1, 4))  # Pick 1 to 4 qualifications
    return ", ".join(selected_qualifications)

def generate_occupation():
    # List of possible occupations
    occupations = [
        "Developer", "Analyst", "Customer Service", "Nursing", "Logistics",
        "Clerk", "Accounting", "Sales", "Manufacturing", "Software Engineer",
        "Product Manager", "Marketing", "HR Specialist", "Data Scientist",
        "Project Manager", "Chef", "Nurse", "Warehouse Manager", "Teacher"
    ]

    # Randomly choose an occupation
    return random.choice(occupations)

def generate_allergy_info():
    has_allergy = random.choice(["Yes", "No"])

    # If the person has allergies, pick random allergies from a predefined list
    if has_allergy == "Yes":
        allergies_list = ["Peanuts", "Dust", "Pollen", "Shellfish", "Latex", "Milk", "Eggs", "Gluten", "Soy", "Wheat", "Mold"]
        selected_allergies = random.sample(allergies_list, random.randint(1, 3))  # Pick 1 to 3 allergies
        return has_allergy, ", ".join(selected_allergies)
    else:
        return has_allergy, "N/A"

def generate_health_condition():
    # Whether to go to the hospital
    hospital_visit = random.choice(["Yes", "No"])

    health_info = {
        "Whether to go to hospital": hospital_visit,
        "Reason for hospital visit": "",
        "Hospital visit schedule": "",
        "Whether or not to take medication": random.choice(["Yes", "No"]),
        "Reason for taking medicine": ""
    }

    # If the person needs to go to the hospital
    if hospital_visit == "Yes":
        reasons_for_hospital = [
            "Routine check-up", "Emergency treatment", "Surgery", "Specialist consultation",
            "Chronic illness management", "Physical therapy", "Pregnancy", "Accident recovery"
        ]
        health_info["Reason for hospital visit"] = random.choice(reasons_for_hospital)
        # Random frequency of hospital visits
        health_info["Hospital visit schedule"] = random.choice(["Once a week", "Once a month", "Once every three months", "As needed"])

    # Whether or not the person takes medication
    if health_info["Whether or not to take medication"] == "Yes":
        reasons_for_meds = [
            "Chronic illness", "Pain management", "Mental health", "Blood pressure control",
            "Diabetes management", "Infection treatment", "Supplements", "Post-surgery recovery"
        ]
        health_info["Reason for taking medicine"] = random.choice(reasons_for_meds)

    return health_info

import random

def generate_commuting_method():
    commuting_methods = ["Car", "Bicycle", "Moped", "Motorcycle", "Walking", "Public transportation"]
    return random.choice(commuting_methods)


def generate_resignation_reason(is_first_job):
    if is_first_job:
        return "First Job"
    return random.choice([
        "Career advancement", "Personal reasons", "Company restructuring", "Relocation", "Other"
    ])

def generate_highest_education():
    return random.choice([
        "Associate Degree", "Bachelor's Degree", "Master's Degree", "Doctorate", "Other"
    ])

def generate_presbyopia(age_group):
    # Presbyopia is more likely in individuals aged 45 and older
    if age_group in ["45-54", "55-60", "Over 60"]:
        return random.choice(["Yes", "No"])
    else:
        return "No"


def generate_dependents():
    dependents_list = [
        "grandfather", "grandmother", "father", "mother", "older brother",
        "older sister", "younger brother", "younger sister", "childcare",
        "Caregiving",
    ]
    return list(set(random.choices(dependents_list, k=random.randint(0, len(dependents_list)))))

## APPROACH 1
def generate_proximity_to_work_location():
    # Placeholder proximity
    commute_type = random.choice(["hours"])  # Distance in miles or commute time in hours
    return f"{random.randint(1, 3)} hours"
```
